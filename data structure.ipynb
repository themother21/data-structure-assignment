```python
#1 discuss string slicinig and provide example



#String slicing in Python refers to extracting a portion of a string by specifying a range of indices. It is a common operation when working with strings, allowing us to retrieve a substring from a larger string. The general syntax for string slicing is:
#string[start:end:step]
'''Parameters:
start: The index from where the slice begins (inclusive).
end: The index where the slice ends (exclusive).
step: The interval between each index. If omitted, the default is 1'''

#basic slicing
text = "Hello, World!"
sliced_text = text[0:5]
print(text[0:5])


#Omitting Start or End
text = "Hello, World!"
sliced_text = text[:5]   # "Hello" (from start to index 4)
sliced_text = text[7:]   # "World!" (from index 7 to the end)
print(text[:5])
print(text[7:])


#Negative Indices: Negative indices count from the end of the string, where -1 represents the last character.
text = "Hello, World!"
sliced_text = text[-6:]  # "World!" (from index -6 to the end)
print(text[-6:])


#Using Step: The step argument allows skipping characters.
text = "Hello, World!"
sliced_text = text[::2]  # "Hlo ol!" (every second character)
sliced_text = text[::-1] # "!dlroW ,olleH" (reversed string)
print(text[::2])
print(text[::-1])


#Summary:
#String slicing provides an easy way to work with substrings in Python.
#You can manipulate the starting point, ending point, and the interval (step) to achieve various results.


```

```python
#2 Explain the keys features of lists  in python

#Lists in Python are one of the most versatile and widely used data structures. They are ordered, mutable, and can hold elements of different types, which makes them a powerful tool for handling collections of data.


#Key Features of Lists in Python:
        
   #Ordered:
#Lists maintain the order of the elements as they are inserted. Each element has a defined position or index starting from 0 for the first element.

my_list = [10, 20, 30]
print(my_list[0])  # 10

   #Mutable:
#Lists are mutable, meaning that elements can be changed, added, or removed after the list has been created.
my_list = [1, 2, 3]
my_list[1] = 20
print(my_list)  # [1, 20, 3]

  #Heterogeneous Elements:
#A list can store elements of different types, such as integers, floats, strings, or even other lists (nested lists).
mixed_list = [1, "Hello", 3.14, [5, 6]]
print(mixed_list)  # [1, "Hello", 3.14, [5, 6]]

  #Dynamic Size:
#Lists can grow or shrink in size. Elements can be added or removed using various list methods.
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # [1, 2, 3, 4]

   #Indexing and Slicing:
#Like strings, lists support indexing (both positive and negative) and slicing to access specific elements or sublists.
my_list = [10, 20, 30, 40, 50]
print(my_list[2])    # 30
print(my_list[1:4])  # [20, 30, 40]

   #List Methods: Lists come with various built-in methods for common operations:
#append(element): Adds an element to the end of the list.
#extend(iterable): Extends the list by appending elements from an iterable.
#insert(index, element): Inserts an element at the specified index.
#remove(element): Removes the first occurrence of an element.
#pop(index): Removes and returns the element at the specified index (or the last element if no index is provided).
#sort(): Sorts the list in place.
#reverse(): Reverses the list in place
my_list = [3, 1, 4, 1, 5]
my_list.sort()
print(my_list)  # [1, 1, 3, 4, 5]

   #Iterating Over Lists:
#You can iterate over the elements of a list using loops.
my_list = ["apple", "banana", "cherry"]
for item in my_list:
    print(item)
# Output:
# apple
# banana
# cherry
 
    #List Comprehensions:
#A concise way to create lists based on existing lists or other iterables.
squares = [x**2 for x in range(5)]
print(squares)  # [0, 1, 4, 9, 16]



#Summary:
#Lists are ordered, mutable, and can contain elements of any data type.
#They support various operations like indexing, slicing, and have built-in methods for adding, removing, and manipulating elements.
#They are highly flexible and widely used for storing collections of data.


```

```python
#3 Desrcibe how to access,modify and delete elements in a list with examples


    #Accessing Elements in a List
#You can access elements in a list using their index. Python uses 0-based indexing, where the first element is at index 0.
my_list = ['apple', 'banana', 'cherry']
print(my_list)
# Accessing the first element
print(my_list[0])  # Output: 'apple'
# Accessing the last element (using negative index)
print(my_list[-1])  # Output: 'cherry'


    #Modifying Elements in a List
#Since lists are mutable, you can change any element by directly assigning a new value to a specific index.
my_list = [10, 20, 30, 40]
print(my_list)
# Modifying the element at index 1
my_list[1] = 200
print(my_list)  # Output: [10, 200, 30, 40]
# Modifying the last element
my_list[-1] = 400
print(my_list)  # Output: [10, 200, 30, 400]


    #Deleting Elements from a List
#You can remove elements from a list using several methods:
     #del statement:
#Deletes an element at a specific index or even the entire list.
my_list = [1, 2, 3, 4, 5]
print(my_list)
# Delete the element at index 2
del my_list[2]
print(my_list)  # Output: [1, 2, 4, 5]
# Delete the entire list
del my_list


    #remove() method:
#Removes the first occurrence of a specific value
my_list = [10, 20, 30, 20, 40]
print(my_list)
# Remove the first occurrence of 20
my_list.remove(20)
print(my_list)  # Output: [10, 30, 20, 40]


    #pop() method:
#Removes and returns the element at a specific index. If no index is provided, it removes the last element by default.
my_list = [1, 2, 3, 4, 5]
print(my_list)
# Remove the element at index 1
removed_element = my_list.pop(1)
print(my_list)         # Output: [1, 3, 4, 5]
print(removed_element)  # Output: 2
# Remove the last element
my_list.pop()
print(my_list)  # Output: [1, 3, 4]


   #clear() method:
#Removes all elements from the list, leaving it empty.
my_list = [1, 2, 3, 4]
print(my_list)
# Clear the entire list
my_list.clear()
print(my_list)  # Output: []



#Summary:
#Accessing: Use indexing (positive or negative) to retrieve specific elements.
#Modifying: Use index assignment to change an element at a specific position.
#Deleting:
#Use del to remove by index.
#Use remove() to delete the first occurrence of a value.
#Use pop() to remove and return an element at a specific index (or the last element by default).
#Use clear() to empty the list entirely.

```

```python
#4 Compare and Constrast tuples and lists with examples 


#Feature	List	Tuple
#Mutability	Mutable (can be changed)	Immutable (cannot be changed)
#Syntax	Square brackets []	Parentheses ()
#Performance	Slower, more memory use	Faster, less memory use
#Use Cases	Dynamic data collections	Fixed, constant collections
#Methods	Many built-in methods (e.g., append(), remove())	Limited methods (count(), index())
#Nested Structures	Supports nesting (lists of lists)	Supports nesting (tuples of tuples)
#Indexing/Slicing	Supports indexing and slicing	Supports indexing and slicing
#Memory Efficiency	Less efficient	More efficient

#Tuples and lists are both data structures in Python that can store multiple items. 
#However, they have some key differences. Here's a comparison between them


      #Mutability
#Tuple: Immutable (cannot be changed after creation).
#List: Mutable (can be modified after creation).
# Tuple (Immutable)
tup = (1, 2, 3)
# Trying to modify it will raise an error
# tup[0] = 10  # Raises TypeError: 'tuple' object does not support item assignment
# List (Mutable)
lst = [1, 2, 3]
lst[0] = 10  # This works fine
print(lst)  # Output: [10, 2, 3]


       #Syntax
#Tuple: Created using parentheses () or by separating items with commas (without brackets).
#List: Created using square brackets []
# Tuple
tup = (1, 2, 3)
print(tup)
# Tuple without parentheses (optional)
tup2 = 1, 2, 3
print(tup2)
# List
lst = [1, 2, 3]
print(lst)


       # Performance
#Tuple: Because tuples are immutable, they are slightly faster than lists.
#List: Lists are slower because of the overhead involved in maintaining mutability.
import timeit
# Timing tuple creation
tuple_time = timeit.timeit("(1, 2, 3)", number=1000000)
# Timing list creation
list_time = timeit.timeit("[1, 2, 3]", number=1000000)
print(f"Tuple creation time: {tuple_time}")
print(f"List creation time: {list_time}")


       #Use Cases
#Tuple: Used when data should not change, such as coordinates or database records.
#List: Used when you need to modify data, such as appending, updating, or removing items.
# Tuple use case (Coordinates)
coordinates = (40.7128, -74.0060)  # New York City coordinates
# List use case (Shopping cart)
shopping_cart = ["apple", "banana", "cherry"]
shopping_cart.append("date")  # Modifying the list
print(coordinates)
print(shopping_cart)


       #Functions
#Tuple: Supports fewer methods (e.g., count(), index()).
#List: Supports more methods (e.g., append(), remove(), sort(), reverse()).
# Tuple methods
tup = (1, 2, 2, 3)
print(tup.count(2))  # Output: 2
print(tup.index(3))  # Output: 3
# List methods
lst = [1, 2, 2, 3]
lst.append(4)
lst.remove(2)
print(lst)  # Output: [1, 2, 3, 4]


      #Hashability
#Tuple: Hashable if all of its elements are hashable, meaning it can be used as a key in dictionaries.
#List: Not hashable, so it cannot be used as a key in dictionaries.
# Tuple as a dictionary key
dict_with_tuple_key = {(1, 2): "value"}
# List cannot be used as a dictionary key
# dict_with_list_key = {[1, 2]: "value"}  # Raises TypeError: unhashable type: 'list'



    #use tuples when you need a collection of items that should remain constant, 
    #and use lists when you expect to modify the data.



```

```python
#5 Describe the key features of sets and provide examples of their use.

#Key Features of Sets in Python
#A set in Python is an unordered collection of unique elements. 
#Unlike lists or tuples, sets do not allow duplicate items and offer various operations like union,intersection 
#and difference that make them useful for tasks related to mathematical set theory.


      #key features
    #Unordered Collection:
#Sets do not maintain the order of elements, unlike lists or tuples.
#You cannot access items by index
my_set = {1, 2, 3, 4}
# Accessing elements by index is not allowed: my_set[0] will raise an error


     #Unique Elements:
#Sets automatically remove duplicate items.
my_set = {1, 2, 2, 3}
print(my_set)  # Output: {1, 2, 3}


     #Mutable:
#While the elements of a set can’t be indexed or changed directly, you can add or remove elements.
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}


     #Set Operations:
#Union (|): Combines elements from two sets.
#Intersection (&): Finds common elements between two sets.
#Difference (-): Finds elements in one set that are not in another.
#Symmetric Difference (^): Finds elements in either set, but not in both.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
# Union
print(set1 | set2)  # Output: {1, 2, 3, 4, 5}
# Intersection
print(set1 & set2)  # Output: {3}
# Difference
print(set1 - set2)  # Output: {1, 2}
# Symmetric Difference
print(set1 ^ set2)  # Output: {1, 2, 4, 5}


        #No Duplicates:
#Sets automatically remove duplicate values when created or modified
my_set = {1, 1, 2, 2, 3}
print(my_set)  # Output: {1, 2, 3}


       #Frozenset:
#An immutable version of a set is called a frozenset. Once created, its elements cannot be changed.
frozen = frozenset([1, 2, 3])
# frozen.add(4) would raise an error since frozenset is immutable
print(frozen)


         #Examples of Use Cases:
     #Removing Duplicates from a List:
#You can convert a list to a set to remove duplicates, as sets automatically enforce uniqueness.
my_list = [1, 2, 2, 3, 4, 4]
unique_set = set(my_list)
print(unique_set)  # Output: {1, 2, 3, 4}


     #Set Operations in Mathematical Problems:
#Set operations like union, intersection, 
#and difference are handy when dealing with groups of items that share or exclude certain characteristics
# Students enrolled in Math and Science
math_students = {"Alice", "Bob", "Charlie"}
print(math_students)
science_students = {"Bob", "David", "Charlie"}
print(science_students)
# Students enrolled in both Math and Science
both_subjects = math_students & science_students
print(both_subjects)  # Output: {"Bob", "Charlie"}


      #Membership Testing:
#Sets are optimized for membership testing (i.e., checking if an item is in a set) 
#and are faster than lists for this purpose.
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(3 in my_set)  # Output: True
print(6 in my_set)  # Output: False



      #Summary:
#Unordered and unique collections.
#Fast membership checks.
#Useful for mathematical set operations like union, intersection, difference.
#Mutable, but with an immutable alternative (frozenset


```

```python
#6 Discuss the use cases of tuples and sets in python programming. 

#In Python programming, tuples and sets are both data structures used to store collections of items,
#but they have different use cases and characteristics. Here’s a detailed discussion of their use cases:


     #Tuples
#A tuple is an immutable, ordered collection of items. 
#Once a tuple is created, it cannot be modified (i.e., you cannot change, add, or remove elements). 
#Tuples are defined using parentheses ()
  #Use Cases for Tuples:
    
    #Data Integrity (Immutability):
#Tuples are ideal when you want to ensure that the data cannot be changed throughout the program. 
#For example, storing constant configuration values or fixed records like geographic coordinates 
#(latitude, longitude)
location = (40.7128, 74.0060)  # Tuple representing a fixed coordinate (NYC)
print(location)


     #Heterogeneous Data:
#Tuples are often used to group data that have different types. 
#For instance, a tuple might hold an integer, a string, and a list all together.
person_info = (30, 'John Doe', ['Engineer', 'Musician'])
print(person_info)


     #Return Multiple Values from Functions:
#A common use of tuples is in functions that need to return more than one value.
def get_person():
    return ('John', 30, 'Engineer')
person = get_person()
print(person)  # ('John', 30, 'Engineer')


     #As Dictionary Keys:
#Since tuples are immutable, they can be used as keys in dictionaries, while lists cannot.
coordinates_dict = {(40.7128, 74.0060): 'New York', (34.0522, 118.2437): 'Los Angeles'}
print(coordinates_dict)


     #Efficient Memory Usage:
#Tuples are more memory-efficient than lists, 
#especially when dealing with a large number of elements that do not need to be modified.

        #Sets
#A set is an unordered collection of unique elements. 
#Sets are mutable and are defined using curly braces {} or the set() function.
#However, a special type of set called a frozenset is immutable.
      #Use Cases for Sets:
    
          #Eliminate Duplicate Values:
#Sets automatically remove duplicate elements, which makes them useful for ensuring uniqueness in a collection.
numbers = [1, 2, 3, 3, 4, 4, 5]
unique_numbers = set(numbers)  # {1, 2, 3, 4, 5}
print(unique_numbers)
      
         #Mathematical Set Operations:
#Sets support operations like union, intersection, and difference, which are common in mathematical computations
set_a = {1, 2, 3}
set_b = {3, 4, 5}
union = set_a | set_b         # {1, 2, 3, 4, 5}
intersection = set_a & set_b  # {3}
difference = set_a - set_b    # {1, 2}
print(union)
print(intersection)
print(difference)
        

        #Membership Testing:
#Sets are optimized for fast membership testing. 
#Checking if an item exists in a set is more efficient than in a list due to the underlying hash table implementation.
fruits = {'apple', 'banana', 'cherry'}
print('apple' in fruits)  # True


        #Unique Elements from Large Datasets:
#When working with large datasets where you want to filter out duplicates, sets are highly efficient.
data = ['apple', 'orange', 'banana', 'apple', 'orange']
unique_data = set(data)  # {'apple', 'orange', 'banana'}
print(unique_data)


         #Unordered Collections:
#Sets are useful when the order of elements is not important, and all you care about is the presence of items.


          #Set Comparisons:
#Sets are useful for comparing groups of items (e.g., finding common items, differences between sets).
employees = {'Alice', 'Bob', 'Charlie'}
managers = {'Bob', 'David'}
common_people = employees & managers  # {'Bob'}
print(common_people)

          #Summary of Differences:
#Order: Tuples are ordered, sets are unordered.
#Mutability: Tuples are immutable, sets are mutable (though frozensets are immutable).
#Duplicates: Tuples allow duplicates, sets do not.
#Usage: Use tuples for fixed-size, ordered data; use sets when uniqueness and fast membership testing are important.
#Both tuples and sets are powerful structures in Python, 
#and their use depends on whether you need order and immutability (tuples)
#or uniqueness and mathematical operations (sets).


```

```python
#7 Desribe how to add,modify and delete items  in a dictionary with examples

#In Python, a dictionary is a mutable data structure that stores key-value pairs. 
#You can easily add, modify, and delete items in a dictionary. 

      #Adding Items to a Dictionary
#You can add a new key-value pair to a dictionary by assigning a value to a new key using the bracket notation []
# Create an empty dictionary
person = {}
# Add key-value pairs
person['name'] = 'Alice'
person['age'] = 25
person['job'] = 'Engineer'
print(person)
# Output: {'name': 'Alice', 'age': 25, 'job': 'Engineer'}
#You can also use the update() method to add multiple key-value pairs at once.
person.update({'city': 'New York', 'hobby': 'Reading'})
print(person)
# Output: {'name': 'Alice', 'age': 25, 'job': 'Engineer', 'city': 'New York', 'hobby': 'Reading'}


       #Modifying Items in a Dictionary
#To modify an item in a dictionary, you can simply assign a new value to an existing key. 
#This will overwrite the current value associated with that key.
# Modify existing key 'age'
person['age'] = 26
print(person)
# Output: {'name': 'Alice', 'age': 26, 'job': 'Engineer', 'city': 'New York', 'hobby': 'Reading'}
#Similarly, you can modify multiple key-value pairs using the update() method.
# Modify 'city' and 'job'
person.update({'city': 'San Francisco', 'job': 'Data Scientist'})
print(person)
# Output: {'name': 'Alice', 'age': 26, 'job': 'Data Scientist', 'city': 'San Francisco', 'hobby': 'Reading'}


        #Deleting Items from a Dictionary
#There are several ways to delete items from a dictionary:
        
    #Using the del statement:
#The del statement removes the item with the specified key.
# Delete the key 'hobby'
del person['hobby']
print(person)
# Output: {'name': 'Alice', 'age': 26, 'job': 'Data Scientist', 'city': 'San Francisco'}

    #Using the pop() method:
#The pop() method removes the item with the specified key and returns its value.
# Remove and return 'city'
removed_city = person.pop('city')
print(removed_city)  # Output: San Francisco
print(person)
# Output: {'name': 'Alice', 'age': 26, 'job': 'Data Scientist'}
       
    #Using the popitem() method:
#The popitem() method removes and returns the last inserted key-value pair as a tuple.
# Remove and return the last inserted item
last_item = person.popitem()
print(last_item)  # Output: ('job', 'Data Scientist')
print(person)
# Output: {'name': 'Alice', 'age': 26}

     #Using the clear() method:
#The clear() method removes all items from the dictionary, leaving it empty.
# Clear all items from the dictionary
person.clear()
print(person)
# Output: {}


   #Summary of Operations:
#Add items: Use [] to add a single item, or update() to add multiple items.
#Modify items: Use [] or update() to modify existing items.
#Delete items: Use del, pop(), popitem(), or clear() to remove items.
#Each of these methods provides flexible ways to manipulate dictionaries depending on the use case


```

```python
#8 Discuss the importance of dictionary keys being immutable and provide examples


#In Python, dictionary keys must be immutable. 
#This means that once a key is used in a dictionary, it cannot be changed. 
#The reason for this restriction lies in how dictionaries are implemented in Python — 
#they use a hash table under the hood. 
#Immutable objects, like strings, numbers, and tuples, can be reliably hashed (assigned a unique identifier for quick lookup), 
#while mutable objects (like lists and dictionaries themselves) cannot be hashed consistently,making them unsuitable as dictionary keys.


    #Why Keys Must Be Immutable:
#Hashing:
#Python dictionaries rely on hash functions to store and retrieve values efficiently. 
#An immutable object’s hash value remains consistent during its lifetime, ensuring the dictionary’s internal hash table functions correctly.
#Mutable objects like lists or dictionaries can change, leading to unpredictable hash values, which would break the dictionary’s lookup mechanism.

#Consistency:
#If dictionary keys could change after they were inserted, it would lead to inconsistencies. 
#The dictionary might not be able to find the value corresponding to a changed key, which would break the integrity of the data structure.

#Efficient Lookup:
#Immutable keys ensure constant-time complexity (O(1)) for lookups.
#Because the hash value remains the same, Python can quickly find the corresponding value without having to scan the entire dictionary.

#Common Immutable Types as Dictionary Keys:
#Numbers (int, float)
#Strings
#Tuples (as long as the tuple itself only contains immutable elements)

#Examples
   # Using Immutable Types as Dictionary Keys:
#Immutable types such as strings, numbers, and tuples are commonly used as dictionary keys.
# Using string keys
person = {'name': 'Alice', 'age': 30}
print(person['name'])  # Output: Alice
# Using number keys
scores = {1: 'Low', 2: 'Medium', 3: 'High'}
print(scores[2])  # Output: Medium
# Using tuple keys (Immutable)
location = {(40.7128, 74.0060): 'New York', (34.0522, 118.2437): 'Los Angeles'}
print(location[(40.7128, 74.0060)])  # Output: New York

   # Attempting to Use Mutable Types as Dictionary Keys:
#Mutable objects like lists and dictionaries cannot be used as dictionary keys. 
#Trying to do so will raise a TypeError.
# Trying to use a list as a key (will raise a TypeError)
#key = [1, 2, 3]
#data = {key: 'Invalid Key'}  # This will cause an error


   #Using Tuples with Mutable Elements as Keys (Fails):
#Even though tuples are immutable, if they contain mutable elements like lists, they can’t be used as dictionary keys.
## Trying to use a tuple containing a list (mutable) as a key
#key = ([1, 2, 3], 'a')
#data = {key: 'Invalid Key'}  # This will cause an error

    #Immutable Tuple as a Key (Success):
#If a tuple contains only immutable objects, it can be used as a key without issue
key = (1, 2, 3)
data = {key: 'Valid Key'}
print(data[key])  # Output: Valid Key


     #Summary of Key Takeaways:
#Dictionary keys must be immutable because they are hashed for efficient lookups.
#Immutable types (e.g., strings, numbers, tuples) can be used as keys, 
#while mutable types (e.g., lists, dictionaries) cannot.
#Attempting to use a mutable object as a key will raise a TypeError 
#because mutable objects do not produce consistent hash values.



```
